```python
# using the pycountry_convert library
# create a country conversion function
# that changes country name to country code
# then changes country code to continent code

def convert(row):
    try:
        #convert country name to country code
        countryCode = pcc.country_name_to_country_alpha2(row.Country, cn_name_format = "default")
        print(f'{countryCode} for {row.Country}')#adding logging so we know where the problem is

        #convert country code to continent code
        continentCode = pcc.country_alpha2_to_continent_code(countryCode)
        print(continentCode)#adding logging so we know where the problem is
        return continentCode
    except:
        return 'XX'#return XX for unknown country code
```

```python
# to apply to dataframe
# create a new column with continent codes

dfQOL_2["Continent"] = dfQOL_2.apply(convert, axis = 1)
```

# Map Continent Codes to Name

After creating the new column that contains the Continent Codes, We changed the Continent Codes to Continent Name for easeir understanding. 

```python
# map continent codes to continent name

# 1st create a dict with continent code as name and continent name as value
# then using .map(dict)

continentNames = {"AF": "Africa",
                 "AS": "Asia",              
                "EU": "Europe",
                "NA": "North America",
                "SA": "South America",
                 "OC": "Oceania"}

dfQOL_2["Continent"] = dfQOL_2["Continent"].map(continentNames)
```
